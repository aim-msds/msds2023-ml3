## Loading Pretrained Models

One of the notable functionality of hugging face is that in enables us to use open-sourced pretrained models which we can use for our own purpose.

```python
from PIL import Image
from transformers import YolosFeatureExtractor, YolosForObjectDetection
```

Using and deploying a pretrained model is straightforward in hugging face.

```python
image_filepath = 'data/counting-sheep/01.jpeg'
image = Image.open(image_filepath)

feature_extractor = YolosFeatureExtractor.from_pretrained('hustvl/yolos-tiny')
model = YolosForObjectDetection.from_pretrained('hustvl/yolos-tiny')

inputs = feature_extractor(images=image, return_tensors="pt")
outputs = model(**inputs)

# model predicts bounding boxes and corresponding COCO classes
logits = outputs.logits
bboxes = outputs.pred_boxes
```

## Interpreting the Outputs

To understand the output of the model, let's first look at the shape of the logits.

```python
logits.shape
```

We can normalize and obtain class probabilities by taking the softmax of the logit.

```python
import torch
```

```python
probs = torch.exp(logits[0]) / torch.exp(logits[0]).sum(axis=1).unsqueeze(1)
```

```python
probs.sum(axis=0)
```

```python
probs.sum(axis=1)
```

```python
probs
```

An easier approach is to use the `post_process` method of the feature extractor of the pretrained model.

```python
# convert outputs (bounding boxes and class logits) to COCO API
target_sizes = torch.tensor([image.size[::-1]])
results = feature_extractor.post_process(outputs, target_sizes=target_sizes)[0]
```

```python
results['scores']
```

```python
results['labels']
```

```python
probs[:, :91].max(axis=1)
```

## Plotting the Model Outputs

Note that the index in the output directly corresponds to the class category in which the pretrained model was trained on. We can retrieve this mapping using the `id2label` attribute of the configuration object of the pretrained model.

```python
model.config.id2label
```

Now, we can repurpose the output of the model for our specialized use-case. The `confidence` parameter sets the sensitivity of the model output.

```python
confidence = 0.95

# Identify objects in the image
objects = {'labels': [], 'scores': [], 'boxes': []}

for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
    box = [round(i, 2) for i in box.tolist()]
    # let's only keep detections with score > 0.9
    if score > confidence:
        label_name = model.config.id2label[label.item()]
        if label_name == 'sheep':
            print(
                f"Detected {label_name} with confidence "
                f"{round(score.item(), 3)} at location {box}"
            )
            objects['labels'].append(label_name)
            objects['scores'].append(round(score.item(), 3))
            objects['boxes'].append(box)
```

Here are the detected objects in the image with their corresponding bounding boxes.

```python
objects
```

Note that the bounding boxes here correspond to the upper left and lower right coordinate of the box `(x1, y1, x2, y2).`

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
fig, ax = plt.subplots()

ax.imshow(image)
for box in objects['boxes']:
    x1, y1, x2, y2 = box
    width = x2 - x1
    height = y2 - y1
    rectangle = Rectangle((x1, y1), width, height, fill=False, ec='tab:red')
    ax.add_patch(rectangle)


ax.axis('off')
```

## Making the Models Scalable

As we scale the model and use it to predict onto several images, we'd have to run over and over again the above code. Thus, to make it scalable, we can define functions to make them reusable.

```python
def get_objects(
    image, model, feature_extractor, confidence=0.95, verbose=False):
    """Return the detected objects of the model given an image

    Parameters
    ----------
    image : PIL image
        Opened PIL image
    model : hugging face PreTrainedModel
        Pretrained model loaded from hugging face
    feature_extractor : hugging face FeatureExtractor
        Corresponding feature extractor of the loaded model
    confidence : float, default=0.95
        The sensitivity of the model, determines which detected
        objects are to be returned
    verbose : bool, default=False
        Whether to print the detected objects

    Returns
    -------
    objects : dict of lists
        Dictionary containing labels of detected objects, corresponding
        scores and bounding boxes
    """
    # Identify objects in the image
    objects = {'labels': [], 'scores': [], 'boxes': []}

    # Get model output scores
    inputs = feature_extractor(images=image, return_tensors="pt")
    outputs = model(**inputs)
    target_sizes = torch.tensor([image.size[::-1]])
    results = feature_extractor.post_process(outputs, target_sizes=target_sizes)[0]    

    for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
        box = [round(i, 2) for i in box.tolist()]

        if score > confidence:
            label_name = model.config.id2label[label.item()]
            if label_name == 'sheep':
                if verbose:
                    print(
                        f"Detected {label_name} with confidence "
                        f"{round(score.item(), 3)} at location {box}")
                objects['labels'].append(label_name)
                objects['scores'].append(round(score.item(), 3))
                objects['boxes'].append(box)

    return objects
```

```python
def get_bbox_dims(bbox):
    """Return the upper left coordinate, width, and height of the bbox"""
    x1, y1, x2, y2 = bbox
    width = x2 - x1
    height = y2 - y1
    xy = (x1, y1)

    return xy, width, height
```

```python
def plot_bboxes(image, bboxes):
    """Plot the image and bounding boxes on a matplotlib Figure"""
    fig, ax = plt.subplots()

    ax.imshow(image)
    for bbox in bboxes:
        xy, width, height = get_bbox_dims(bbox)
        rectangle = Rectangle(xy, width, height, fill=False, ec='tab:red')
        ax.add_patch(rectangle)

    ax.axis('off')

    return fig, ax
```

```python
def plot_sheep_boxes(img_filepath, model, feature_extractor, confidence):
    """Plot the bounding boxes of detected sheep objects in the image

    Parameters
    ----------
    img_filepath : str or Path
        The filepath of the image of interest
    model : hugging face PreTrainedModel
        Pretrained model loaded from hugging face
    feature_extractor : hugging face FeatureExtractor
        Corresponding feature extractor of the loaded model
    confidence : float, default=0.95
        The sensitivity of the model, determines which detected
        objects are to be returned

    Returns
    -------
    objects : dict of lists
        Dictionary containing labels of detected objects, corresponding
        scores and bounding boxes

    fig, ax : matplotlib Figure and Axes
        Figure and axes of the generated plot
    """
    image = Image.open(img_filepath)
    objects = get_objects(image, model, feature_extractor, confidence)
    fig, ax = plot_bboxes(image, objects['boxes'])

    return objects, fig, ax
```

Let's now use our function to plot the bounding boxes of the detected sheep in several images.

```python
img_filepath = './data/counting-sheep/01.jpeg'

plot_sheep_boxes(img_filepath, model, feature_extractor, 0.95);
```

```python
img_filepath = './data/counting-sheep/02.jpeg'

plot_sheep_boxes(img_filepath, model, feature_extractor, 0.90);
```

```python
img_filepath = './data/counting-sheep/03.jpeg'

plot_sheep_boxes(img_filepath, model, feature_extractor, 0.85);
```
